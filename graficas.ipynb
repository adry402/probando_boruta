```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline #Para que imprima inline en el notebook
rcParams['figure.figsize'] = 5,4 # Dimensiones del plot  5 para width, 4 para top
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure() # Se crea la figura
ax = fig.add_axes([.1,.1,1,1]) # se crean los axes

#Dibujamos

ax.plot(x,y)
```

```python
#se agregan limites
fig = plt.figure() 
ax = fig.add_axes([.1,.1,1,1]) 

ax.set_xlim([1,9])
ax.set_ylim([0,5])

#tickmarks

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```python
# add grid

fig = plt.figure() 
ax = fig.add_axes([.1,.1,1,1]) 

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

```python
# multiple plot

fig = plt.figure()
fig, (ax1,ax2) = plt.subplots(2,1) #1 row, 2 columns

ax1.plot(x)
ax2.plot(x,y)
```

```python
z = [1,2,3,4,0.5]
veh_type = ['bycicle','motorbyke','car','van','stroller']

plt.pie(z, labels=veh_type)
plt.show()
```

```python
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg =  cars.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation = 60, fontsize = 'medium')
ax.set_title('Miles per gallon of cars in mtcars dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles per gallon')
ax.grid()
```

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```
